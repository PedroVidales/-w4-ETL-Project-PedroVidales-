### 2.Unimos los DataFrames dogs y dogin en funci√≥n de la columna Breed en un nuevo DataFrame llamado result.

```python
common_breeds = list(set(dogs['Breed']).intersection(dogin['Breed']))
common_breeds
```

```python
dogs_filtered = dogs[dogs['Breed'].isin(common_breeds)]
dogin_filtered = dogin[dogin['Breed'].isin(common_breeds)]
```

```python
result = pd.merge(dogs_filtered, dogin_filtered, on='Breed')
```

```python
result = result.sort_values(by='reps_lower', ascending=True) #Ordenamos el DataFrame en funci√≥n de la columna reps_lower.
```
