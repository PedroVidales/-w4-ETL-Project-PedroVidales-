```python
dogs = pd.read_csv('../data/archivo/dog_breeds.csv', encoding = "ISO-8859-1")
dogin = pd.read_csv('../data/archivo/dog_intelligence.csv', encoding = "ISO-8859-1")
```

```python
dogs = dogs.drop('Height (in)', axis=1)
```

```python
dogs.columns
```

```python
common_breeds = list(set(dogs['Breed']).intersection(dogin['Breed']))
common_breeds
```

```python
dogs_filtered = dogs[dogs['Breed'].isin(common_breeds)]
dogin_filtered = dogin[dogin['Breed'].isin(common_breeds)]
```

```python
result = pd.merge(dogs_filtered, dogin_filtered, on='Breed')
```

```python
result = result.sort_values(by='reps_lower', ascending=True)
```
